```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn

data1 = pd.read_csv("./population-by-single-year-of-age-and-sex-for-nepal-2011.csv", skiprows=2)

data2 = data1.nlargest(10, 'Total')

x = data2['Age']
y = data2['Total']

plt.figure(figsize=(12,8))
plt.title("Top 10 populated age group population in Nepal")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.xticks(rotation=45, ha='right')

plt.bar(x, y, color='maroon', width=0.4)

plt.show()
```
