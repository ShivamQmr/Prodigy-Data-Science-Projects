```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn

data1 = pd.read_csv("./world-bank-data/API_SP.POP.TOTL_DS2_en_csv_v2_144171.csv", skiprows=4)

data2 = data1.nlargest(10, '1960')

x = data2['Country Name']
y = data2['1960']

plt.figure(figsize=(12,8))
plt.title("Population of top most populated countries/region in 1960")
plt.xlabel("Country/Region Name")
plt.ylabel("Population in Billions")
plt.xticks(rotation=45, ha='right')

plt.bar(x, y, color='maroon', width=0.4)

# fig, ax = plt.subplots()
# ax.plot([1,3,5], [3,5,6])
plt.show()
```
